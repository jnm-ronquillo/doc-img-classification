```python
folder_path = 'data'
output_path = 'data_oversampled'
```

```python
df.columns
```

```python
for row in df.itertuples(index=False):
    file_path = os.path.join(folder_path, row.filename)
    ext = Path(row.filename).suffix
    file_name = Path(row.filename).stem
    destination_name = f'{file_name}_{row.duplication_number}{ext}'
    sub_folder_path = os.path.join(output_path, row.category)    
    destination_path = os.path.join(sub_folder_path, destination_name)
    print(destination_path)    
    os.makedirs(sub_folder_path, exist_ok=True)
    shutil.copy2(row.filepath, destination_path)
    
```
